```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import os
import time
#os.chmod('./chromedriver', 755)
```

```python
chrome_options = Options()
chrome_options.add_argument("--incognito")
chrome_options.add_argument("--window-size=1920x1080")
```

```python
driver = webdriver.Chrome('./chromedriver')
url = "https://www.dropbox.com/sh/qe9h0jdh7fgz006/AADWQ5v8_4YRIF61uAbu75oua?dl=0&lst=&preview=horarios_20211_Facultad_Artes.pdf"
driver.get(url)
time.sleep(3)
element = driver.find_element_by_class_name('viewer-container')
item2 = element.find_element_by_class_name("EQJE7qmsBz")
item3 = item2.find_element_by_tag_name('ol')
item4 = item3.find_elements_by_tag_name('li')

for data in item4:        
    if int(data.get_attribute("data-index")) == 10:
        break                                    
data.location_once_scrolled_into_view

def proceso(paginas):    
    item5 = item3.find_elements_by_tag_name('li')    
    for data2 in item5:            
        if int(data2.get_attribute("data-index")) > paginas:
            print("pagina: " + str(paginas) + " dato: " +data2.get_attribute("data-index"))                
    data2.location_once_scrolled_into_view    

paginas = 0
while paginas <= 410:#cantidad de paginas del pdf        
    #h +=2
    #print(h)
    if paginas == 0:                
        proceso(0)
        time.sleep(3)    
        paginas += 12        
    else:        
        proceso(paginas)
        paginas += 2
        time.sleep(3)    
    paginas += 11
```
